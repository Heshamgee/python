<span style="font-size: 14px;font-weight:bold;color: black;">methods on list </span>
1. append(x): Adds an element x to the end of the list.

2. extend(iterable): Extends the list by appending elements from the iterable.

3. insert(i, x): Inserts element x at the specified index i.

4. remove(x): Removes the first occurrence of element x from the list.

5. pop([i]): Removes and returns the element at index i. If i is not provided, removes and returns the last element.

6. index(x): Returns the index of the first occurrence of element x.

7. count(x): Returns the number of occurrences of element x in the list.

8. sort(): Sorts the elements of the list in ascending order.

9. reverse(): Reverses the order of the elements in the list.

10. copy(): Returns a shallow copy of the list.

11. sum(iterable): Returns the sum of elements in the iterable (not specific to lists, but commonly used).

<span style="font-size: 14px;font-weight:bold;color: black;">built in function on list </span>

1. len(iterable): Returns the number of items in the iterable 

2. sum(iterable, start=0): Returns the sum of all elements in the iterable.

3. min(iterable): Returns the smallest element in the iterable.

4. max(iterable): Returns the largest element in the iterable.

5. sorted(iterable, key=None, reverse=False): Returns a new sorted list from the elements in the iterable.

6. any(iterable): Returns True if at least one element of the iterable is True. If the iterable is empty, returns False.

7. all(iterable): Returns True if all elements of the iterable are True. If the iterable is empty, returns True.

8. enumerate(iterable, start=0): Returns an enumerate object that produces tuples containing indices and values from the iterable

```python
#list and tuple = variable can hold many values with diff data type like array and its zero index
#list is mutable means can change values after created (add,remove,change)
#tuple is unmutable and slightly smaller memory than lis
x=[5,9,78,53]
print(x)
type(x)
print(x[0])
print(x[1])
#change a value of list on index 0
x[0]=80
print(x)
#matrix = is a list 2D
mat =[[4,7],["g","q"]]
print(mat)
#Method on list 
#append list
num =[1,4,8]
print(num)
num.append(9)
print(num)
print(x)
num.extend(x)
print(num)
```

<span style="font-size: 18px; font-weight:bold;color: blue;">Loops </span>
loop is an instruction that repeats multible times as the condition met
1. while loop : 
   is used to repeatedly execute a block of code as long as a given condition is true. 

2. for loop :
    Use for: Known sequences, specific number of iterations (list,tuple,string,array)


```python
#while loop 
var = 14

while var % 7 != 0 :
     var = int(input("Enter the number multible 7"))
else :
    print(f"the number %s is multible of 7 " %var )
    
```

```python
#for loop 
x =[[1,2,3],["a","b","c"]]
for i in x :
    for j in i :
        print(i)
```
